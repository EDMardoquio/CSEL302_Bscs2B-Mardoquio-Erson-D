```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('Lesson_4_student_grades_dataset.csv')
data.head()
```

```python
plt.scatter(data['Hours_of_Study'],data['Grades'])
plt.xlabel('Independent Variable {x}')
plt.ylabel('dependent Variable {y}')
plt.title('Scatter plot of Data')
plt.show()
```

```python
x = data[['Hours_of_Study']]
y = data[['Grades']]
```

```python
model = LinearRegression()

model.fit(x,y)
```

```python
slope = model.coef_[0]

intercept = model.intercept_

print('slope',slope)
print('Intercept',intercept)
```

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x),color=('red'))
plt.xlabel('Independent Variable {x}')
plt.ylabel('dependent Variable {y}')
plt.title('Linear Regression')
plt.show()
```

```python
new_x = [[np.random.randint(100,200)]]

predicted_y = model.predict(new_x)

print('Value of x: ',new_x)
print('Predicted Value of y: ',predicted_y)
```
